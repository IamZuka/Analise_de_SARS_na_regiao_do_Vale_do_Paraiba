```python
import pandas as pd
low_memory=False
%matplotlib inline
pd.options.display.max_columns = 100000
pd.options.display.max_rows = 80000
```

```python
filename = r'C:\Users\Zuka\Documents\Analise de SARS no vale do paraiba\INFLUD-07-07-2020.csv'
data = pd.read_csv(filename, delimiter=';', encoding='ISO-8859-1',error_bad_lines=False)
```

```python
data.shape
```

```python
data.info()
```

```python
cod_etinia = {1:'Branca', 2:'Negro', 3:'Asiatico', 4:'Parda', 5:'Indígena', 9:'Ignorado'}
cod_escolaridade = {0:'Sem', 1:'Fund1', 2:'Fund2', 3:'Médio',4:'Superior',5:'Não se aplica',9:'Ignorado'}
#cod_sexo = {1:'Masculino', 2:'Feminino', 9:'Ignorado'}
cod_evolucao = {1:'Curados', 2:'Óbitos', 9:'Ignorado'}
cod_zona = {1:'Urbana', 2:'Rural', 3:'Periurbana', 9:'Ignorado'}
```

```python
data.loc[(data.CS_ZONA.isnull()), 'CS_ZONA'] = 9
data['CS_ZONA'] = data['CS_ZONA'].apply(lambda x: cod_zona[x])
data.loc[(data.CS_RACA.isnull()), 'CS_RACA'] = 9
data['CS_RACA'] = data['CS_RACA'].apply(lambda x: cod_etinia[x])
data.loc[(data.CS_ESCOL_N.isnull()), 'CS_ESCOL_N'] = 9
data['CS_ESCOL_N'] = data['CS_ESCOL_N'].apply(lambda x: cod_escolaridade[x])
data.SEM_NOT = data.SEM_NOT.apply(pd.to_numeric, errors='coerce')
data['DT_NOTIFIC'].dropna(inplace=True)
```
